```scala
val typeMap = train.dtypes.map(column => 
    column._2 match {
        case "IntegerType" => (column._1 -> 0)
        case "StringType" => (column._1 -> "0")
        case "DoubleType" => (column._1 -> 0.0)
    }).toMap
```

```scala
val train2 = train.na.fill(typeMap)
```

```scala
train2.show(numRows = 6, truncate = 200, vertical=true)
```

```scala
train2.printSchema
```
